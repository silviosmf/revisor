# Analise inicial da base sem duplicados

Este notebook carrega `base_sem_duplicados.csv` e apresenta uma visao inicial da base para orientar as proximas analises.

```python
from pathlib import Path

import pandas as pd
from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 120)

repo_root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data_path = repo_root / 'dados' / 'identificados' / 'base_sem_duplicados.csv'

df = pd.read_csv(data_path)
# df = df[df['publication_year'] > 2014]
print(f'Arquivo: {data_path}')
print(f'Linhas: {df.shape[0]:,}'.replace(',', '.'))
print(f'Colunas: {df.shape[1]}')
```

```python
resumo_colunas = pd.DataFrame({
    'tipo': df.dtypes.astype(str),
    'nao_nulos': df.notna().sum(),
    'nulos': df.isna().sum(),
    'percentual_nulos': (df.isna().mean() * 100).round(2),
    'valores_unicos': df.nunique(dropna=True)
}).sort_values(['percentual_nulos', 'valores_unicos'], ascending=[False, False])

display(resumo_colunas)
```

```python
anos = df['publication_year'].value_counts().sort_index().rename_axis('publication_year').to_frame('quantidade')
display(anos.tail(20))
```
